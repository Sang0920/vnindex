```python
from tradingview_screener import Query, col

# 1. Create a query for the 'vietnam' market
# 2. Select the columns you want (Ticker, Name, Close Price, Volume, etc.)
# 3. Filter specifically for the 'HOSE' exchange
# 4. Limit to a high number (e.g., 600) to ensure you get all constituents
query = (Query()
         .set_markets('vietnam')
         .select('name', 'close', 'volume', 'market_cap_basic', 'price_earnings_ttm', 'sector')
         .where(col('exchange') == 'HOSE')
         .limit(9999)
         .get_scanner_data())

# The result 'query' is a tuple: (total_count, pandas_dataframe)
count, df = query

# Display the data
print(f"Found {count} stocks on VNINDEX (HOSE)")
print(df.head())

# Optional: Export to CSV
# df.to_csv('vnindex_constituents.csv', index=False)
```
